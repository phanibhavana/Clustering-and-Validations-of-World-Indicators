**K-Means**

**Silhouette score**

```python
# Fornula for silhouette score
#    a(i) = (1/|Ca|)   Î£          d(i,j)
#                     jâˆˆCa,iÌ¸=j
#    The average distance a(i) to all the examples in the same cluster is computed
#    b(i) = min (1/|Cb|)   Î£          d(i,j)
#           Cbâ‰ Ca         jâˆˆCb
#   The minimum average distance b(i) between the example and the examples contained in each cluster not containing the analyzed example
#   s(i) = b(i) âˆ’a(i)/max{a(i),b(i)}

          
```

```python
range_n_clusters = [2, 3, 4, 5, 6]

for n in range(2, 7):
    
    clusterer = KMeans(n_clusters=n, random_state=10)
    cluster_labels = clusterer.fit_predict(world_df.iloc[:,0:16])

    silhouette_avg = silhouette_score(world_df.iloc[:,0:16], cluster_labels)
    print("For n_clusters =", n, "The average silhouette_score is :", silhouette_avg)
```

Here, the silhouette values show that the n_clusters value of 4, 5 and 6 are a bad pick for the given data due to the presence of clusters with low average silhouette scores.

```python
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3,4,5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=10)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(world_df.iloc[:,0:16])
```

The silhouette plot for n_clusters=3 is comparatively low than n_clusters=2 and also the thickness and size of clusters are uneven in n_clusters=3 and more accurate and sensible in n_clusters=2. So, we can consider 2 clusters for the given data for kmeans. (k=2)

```python
means = []
interia = []
for i in range(1, 10):

    kmeans = KMeans(n_clusters=i)
    kmeans.fit(world_df.iloc[:,0:16])
    interia.append(kmeans.inertia_)
    means.append(i)
    
    print('SSE Data: '+ str(kmeans.inertia_))
```
